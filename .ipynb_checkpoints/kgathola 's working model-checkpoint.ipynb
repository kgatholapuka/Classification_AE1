## **Preprocess the Data**

### **Text Preprocesing**

```python
# checking for noise in the dataset
df['message'].apply(lambda x: nt.TextFrame(x).noise_scan())
```

```python
# exploring the noise words and the thnhe stopwords
df['message'].apply(lambda x:nt.TextExtractor(x).extract_stopwords())
```

```python
message = df['message']
from nltk.corpus import stopwords

# removing the stopwords

stop_words = set(stopwords.words('english'))
message = message.str.lower()
message = message.apply(lambda x: ' '.join(term for term in x.split() if term not in stop_words))
```

```python
# Replace email addresses with 'email'
processed = message.str.replace(r'^.+@[^\.].*\.[a-z]{2,}$',
                                 'emailaddress')

# Replace URLs with 'webaddress'
processed = processed.str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$',
                                  'webaddress')

# Replace money symbols with 'moneysymb' (£ can by typed with ALT key + 156)
processed = processed.str.replace(r'£|\$', 'moneysymb')
    
# Replace 10 digit phone numbers (formats include paranthesis, spaces, no spaces, dashes) with 'phonenumber'
processed = processed.str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{4}$',
                                  'phonenumbr')
    
# Replace numbers with 'numbr'
processed = processed.str.replace(r'\d+(\.\d+)?', 'numbr')
```

```python
#Remove punctuations
# Remove punctuation
processed = processed.str.replace(r'[^\w\d\s]', ' ')

# Replace whitespace between terms with a single space
processed = processed.str.replace(r'\s+', ' ')

# Remove leading and trailing whitespace
processed = processed.str.replace(r'^\s+|\s+?$', '')
```

```python
#removing the word stem using porter stemmer
ps = nltk.PorterStemmer()

processed = processed.apply(lambda x: ' '.join(ps.stem(term) for term in x.split()))
```

```python
#word cloud
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
#% matplotlib inline
wordcloud = WordCloud(background_color='pink',stopwords=stop_words,max_words=100,max_font_size=50, random_state=42).generate(str(processed))
print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
from nltk.tokenize import word_tokenize

# create a bag_of_words
all_word = []

for messages in processed:
    word = word_tokenize(messages)
    for w in word:
        all_word.append(w)
    
all_word = nltk.FreqDist(all_word)    
```

```python
print('Number of words: {}'.format(len(all_word)))
print('Most common words: {}'.format(all_word.most_common(15)))
```

```python
#geting the most frequent words in the text
word_features = list(all_word.keys())[:1500]
```

```python
#Convert most frequent words to dataframe for ploting the bar plot
def find_features(mess):
    words = word_tokenize(mess)
    feature = {}
    for word in word_features:
        feature[word] = (word in words)
    return feature

#lets see an example of what we just did
feature = find_features(processed[0])
for key,value in feature.items():
    if value == True:
        print(key)
```
